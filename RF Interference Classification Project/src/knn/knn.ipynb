```python
## this is painfuly slow due to size of dataset. and very inaccurate. Maybe try weighing the 
## minority classes.
import numpy as np
import torch.nn as nn
from my_dataset import MyDataset 
from voting_model import MyVoter, MyTrainedModel, MyNN, MyLinear
import matplotlib.pyplot  as plt

train_filepath = "my_data/superfeatures1/train.csv"
test_filepath = "my_data/superfeatures1/test.csv"

train_data = np.loadtxt(train_filepath, delimiter=",", skiprows=1)
test_data  = np.loadtxt(test_filepath,  delimiter=",", skiprows=1)

train_y = train_data[:, -1].astype(int)
test_y = test_data[:, -1].astype(int)

train_X1 = train_data[:, :-1]
test_X1  = test_data[:, :-1]

X = train_X1
y = train_y



```

```python
k = 10
def knn_eval(x):
    sq_X = np.power(X - x, 2)
    ssq_X = np.sum(sq_X, axis=1)
    sssq_X = np.sqrt(ssq_X)
    # print(ssq_X.shape)
    # print(sssq_X.shape)

    min_idx = np.argsort(sssq_X, axis=0)
    # print(min_idx)
    min_labels = y[min_idx]
    k_labels = min_labels[0:k]
    counts = np.bincount(k_labels, minlength=9)
    return counts #(num_classes)

# test_knn: (num_test_samples, num_classes)
test_knn = []
correct = 0
i = 0
for x, lbl in zip(test_X1,test_y):
    eval = knn_eval(x)
    test_knn.append(eval)
    # print(eval, " - ", lbl)
    if np.argmin(eval) == lbl: correct += 1
    print(f"{i} / {len(test_y)}, acc={correct/(i+1)}", end="\r"); i+=1


```
